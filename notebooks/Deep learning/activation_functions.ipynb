# ACTIVATION FUNCTIONS

## Importing libraries

```python
import matplotlib.pyplot as plt
import numpy as np
```

### SIGMOID


```python
def sigmoid(x):
    return 1/(1+np.exp(-x))

## Plotting
x_value = np.linspace(-10, 10, 100)
plt.plot(sigmoid(x_value))
plt.title("Sigmoid function")
plt.show()
```

### RELU

```python
def relu(x):
    return np.maximum(0,x)

## Plotting
x_value = np.linspace(-10, 10, 100)
plt.plot(x_value,relu(x_value))
plt.title("RELU function")
plt.show()
```

## LEAKY RELU

```python
def l_relu(x, alpha=0.1):
    return np.maximum(alpha*x,x)

## Plotting
x_value = np.linspace(-10, 10, 100)
plt.plot(l_relu(x_value))
plt.title("LEAKY RELU function")
plt.show()
```

### SOFTMAX

```python
def softmax(x):
    exp = np.exp(x)
    return exp/np.sum(exp)

## Plotting
x_value = np.linspace(-10, 10, 100)
plt.plot(softmax(x_value))
plt.title("SOFTMAX function")
plt.show()
```
