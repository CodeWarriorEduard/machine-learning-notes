### Using numpy

```python
a = np.array([1,2,3]) # one-dimensional array
b = np.array([4,5,6])
print("shape: \n", a.shape)
print("a+b", a+b)
print("a*b", a*b)
```

```python
# Asignamos un vector de dos dimensiones
a = np.array([[1,2],[3,4]]) # Multidimensional array (2,2)
print(a.shape)
print("a:")
print(a)
print('a.sum(axis=0):', a.sum(axis=0))
print('a.sum(axis=1):', a.sum(axis=1))
```

## SQUARE FUNCTION

```python
def square(x: np.ndarray) -> np.ndarray:
    return np.power(x,2)

input_values = np.arange(-10,10,0.001)
sq = square(input_values)
print(sq)
plt.title('Square function')
plt.plot(input_values,sq)
```

## LEAKY RELU FUNCTION

```python
def l_relu(x: np.ndarray) -> np.ndarray:
    return np.maximum(0.2*x,x)

l_r = l_relu(np.array([1,6,3]))
input_values = np.arange(-10,10,0.001)
l_r = l_relu(input_values)
print(sq)
plt.title('LEAKY RELU function')
plt.plot(input_values,l_r)
print(l_r)
```
