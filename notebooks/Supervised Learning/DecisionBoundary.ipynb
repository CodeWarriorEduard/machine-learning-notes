```python
import numpy as np
import matplotlib.pyplot as plt
```

### Declare a basic dataset.

```python
X = np.array([[0.5, 1.5], [1,1], [1.5, 0.5], [3, 0.5], [2, 2], [1, 2.5]])
y = np.array([0,0,0,1,1,1]).reshape(-1,1)

y = y.reshape(-1,1)

X.shape, y.shape


print(X)
print(y)
```

### Plot data


```python
fig, ax = plt.subplots(1,1,figsize=(4,4))

y_flat = y.flatten()

ax.scatter(X[:,0], X[:,1], c=y_flat, cmap='bwr')

ax.axis([0,4,0,3.5])
ax.set_ylabel('x_1')
ax.set_xlabel('x_0')


plt.show()
```

```python
### PLOTTING DECISION BOUNDARY

x0 = np.arange(0,6)
x1 = 3 - x0


print(x0)
print(x1)

fig, ax = plt.subplots(1,1, figsize=(5,4))

ax.plot(x0,x1, c="b")
ax.axis([0,4,0,3.5])

ax.fill_between(x0,x1, alpha=0.2)


y_flat = y.flatten()

ax.scatter(X[:,0], X[:,1], c=y_flat, cmap='bwr')

ax.axis([0,4,0,3.5])
ax.set_ylabel('x_1')
ax.set_xlabel('x_0')


```
